# Nötprovning 2024-09-13

## Introduktion

Nötter är livsmedel som ofta intas som mellanmål eller tilltugg. Rostade, saltade nötter av olika fabrikat varierar i bland annat smak, konsistens, pris och upplevd smaklighet. Denna rapport redogör för en blindprovning av olika fabrikat av nötter i syfte att utröna:

- Vilka skillnader i smak och konsistens kan testpersoner identifiera mellan olika fabrikat av nötter (cashewnötter, jordnötter, pistagenötter)?
- Hur samvarierar dessa egenskaper med testpersonernas preferenser mellan olika fabrikat?

## Försöksdesign

Nötter blindprovades med tolv testpersoner 2024-09-13.

### Nötter och dryck

Det fanns tre provningsstationer med olika sorters nötter:

- Cashewnötter (6 fabrikat)
- Jordnötter (6 fabrikat)
- Pistagenötter (5 fabrikat)

Vid varje provningsstation serverades de olika fabrikaten ur likadana papptallrikar märkta med koder C1, C2, ..., C6 (cashewnötter), J1, ..., J6 (jordnötter) och P1, ..., P5 (pistagenötter). Identiteten på de olika nötterna (C1 = Den lille nöttefabrikken, C2 = Alesto, etc.) avslöjades för testpersonerna först efter provningen.

De tolv testpersonerna fick prova nötterna i valfri ordning. Under provningen serverades olika drycker (vatten, öl och bål på klarbärsvin) att inta ad libitum.

### Uppmätta parametrar

Deltagarna fyllde i ett provningsformulär där fyra beskrivande parametrar och ett helhetsbetyg värderades på en heltalsskala från 1 till 7. Parametrarna var:

- Flottig (1-7)
- Knaprig (1-7)
- Rostad (1-7)
- Salt (1-7)
- Betyg (1-7)

Testpersonerna kunde dessutom lämna skriftliga kommentarer om de olika nötterna.

### Pseudonym datainsamling

Varje testperson tilldelades ett provningsformulär i form av ett A4-papper med instruktioner och formulär för utvärdering. Provningsformuläret finns att tillgå i bilagd med PDF-fil.

Varje utvärderingsformulär märktes med en unik pseudonym kod (trebokstavsförkortning, till exempel AMK, ARN, eller FHM) som alltså svarar mot testpersonens identitet.

## Dataanalys

En översiktlig dataanalys har genomförts enligt följande. Nötternas egenskaper (flottig, knaprig, rostad, salt) kvantifierades på gruppnivå som medelvärde av testpersonernas svar. Testpersonernas genomsnittliga helhetsbetyg räknades också ut.

Approximativa konfidensintervall (95%) för nötternas egenskaper och totalbetyg beräknades som

$$
\text{CI} = \mu \pm 1.96 \frac{\sigma}{\sqrt{n}},
$$

där $\mu$ är genomsnittsvärdet, $\sigma$ är standardavvikelse och $n$ antalet svar.

För att undersöka hur väl nötternas egenskaper predikterar helhetsbetyget skattades en enkel regressionsmodell separat för varje egenskap:

$$
\text{betyg} = \beta_0 + \beta_1 \cdot \text{egenskap}.
$$

De statistiska modellerna passades med testpanelens medelvärden (inte individuella data). Statistisk signifikans på regressionsparametern $\beta_1$ bedömdes utifrån p-värde i denna regressionsanalys.

**Figur 1** Resultat för cashewnötter, jordnötter och pistsgenötter. De olika panelerna visar hur helhetsbetyget samvarierar med egenskaperna flottig, knaprig, rostad och salt. Felmarkörerna kring punkterna visar 95-procentiga konfidensintervall för testpanelens genomsnittliga värdering av egenskaper och helhetsbetyg. Regressionslinjer är inritade och angivna p-värden gäller regressionslinjens lutning.

Helhetsbetyget på jordnötter samvarierade tydligt med nötternas egenskaper medan helhetsbetyg på övriga nötter har ett mindre tydligt förhållande till egenskaperna.

- Jordnötternas helhetsbetyg varierade kraftigt mellan fabrikaten. Helhetsbetyget samvarierade starkt och positivt framförallt med egenskaperna salt och rostad ($p < 0.01$). En viss positiv samvariation kunde möjligen skönjas även med egenskapen knaprig ($p=0.09$). Ett av fabrikaten (ICA I love eco) stack ut med mycket lägre helhetsbetyg än övriga jordnötter och samtidigt lägst värdering av alla jordnötter i egenskaperna knaprig, rostad och salt. Några testpersoner lämnade nedsättande kommentarer om detta fabrikat på formulären.
- Cashewnötternas helhetsbetyg varierade mindre än jordnötternas. Helhetsbetyget samvarierade med egenskapen knaprig ($p=0.01$).
- Pistagenötternas helhetsbetyg varierade relativt lite och samvarierade dessutom inte signifikant med någon av egenskaperna.

Det är inte möjligt att säga vilket orsakssamband som finns mellan nötternas egenskaper och helhetsbetyg, även om det verkar troligt att de uppmätta egenskaperna är delorsaker till varför testpersonernas helhetsbetyg varierar. För jordnötterna är egenskaperna knaprig, salt och rostad även starkt positivt korrelerade med varandra och det går inte att säga vilka av egenskaperna som är viktigast för helhetsbetyget. Det kan förstås inte heller uteslutas att det finns andra, icke kvantifierade, egenskaper som avgör nötternas betyg. För att reda ut orsakssamband skulle det krävas ytterligare undersökningar, t ex med ett bredare urval av nötter där egenskaperna varierar mer, och/eller en kvalitativ undersökning där testpersonerna får ange varför de gillar vissa nötter mer än andra. Vissa initiala ledtrådar kan finnas i de kommentarer som några testpersoner har lämnat om olika nötter, men detta utvärderas inte i denna rapport.

Notera att dataanalysen är genomförd på gruppnivå, alltså utifrån testpanelens genomsnittliga värderingar. Individuella variationer i upplevelse kan avvika från dessa resultat på gruppnivå. En mixed effects-modell eller annan mer sofistikerad statistisk design skulle vara av intresse men har inte provats. Ytterligare undersökningar kan göras i det bifogade datamaterialet.

Att de uppmätta egenskaperna (flottig, knaprig, rostad, salt) varierade kraftigast för jordnötter är ett väntat resultat eftersom de fyra egenskaperna valdes särskilt för jordnötter i en tidigare provning (https://github.com/rasmuse/peanuts). Till en utökad provning av cashew- och pistagenötter skulle det vara lämpligt att överväga andra egenskaper att kvantifiera. Författarens personliga mening är dock att olika fabrikat av jordnötter verkligen varierar mer än andra nötter i smakprofil och konsistens och att det därför är tveksamt om en annan uppsättning av egenskaper skulle kunna ge någon väsentligt förbättrad förståelse för preferens mellan olika fabrikat av cashew- och pistagenötter.

En reflektion om testningen är att flera testpersoner upplevde det svårt och krävande att rättvist gradera fem till sex sorters likartade nötter i så många aspekter. Flera testpersoner rapporterade att de olika sorterna (särskilt cashew- och pistagenötter) till slut upplevdes smaka ungefär likadant. En sorts nöttrötthet infann sig, som dock gick att uthärda tack vare gott sällskap och ovan nämnda drycker. För att undvika nöttrötthet i framtida provningar kan det vara aktuellt att använda en annan försöksdesign, till exempel där testpersonerna får prova slumpmässigt valda par av olika nötter och utvärdera jämföra egenskaper och helhetsbetyg som parvisa jämförelser. En sådan försöksdesign övervägdes till denna provning men valdes bort eftersom den är logistiskt mer komplicerad att genomföra och eftersom det möjligen också blir svårare att nå statistiskt signifikanta resultat med en liten testpanel.

## Slutsatser

Testpanelens genomsnittliga helhetsbetyg på jordnötter 
varierade kraftigt och samvarierade positivt med egenskaperna knaprig och salt. Cashewnötternas helhetsbetyg samvarierade positivt med egenskapen knaprig. Pistagenötterna hade inga större skillnader i genomsnittligt helhetsbetyg eller egenskaper.

På gruppnivå blir slutsatsen att valet av jordnötter stor roll för helhetsbetyget medan valet av cashew- och pistagenötter spelar mindre roll.

Fortsatt forskning kan bidra med ytterligare insikter. Bland annat kan kvalitativa och kvantitativa studier bidra till att klargöra orsakssamband mellan egenskaper och helhetsbetyg. Individuella variationer i gradering av egenskaper och helhetsbetyg är också intressant föremål för ytterligare studier och en initial sådan studie kan redan göras med datamaterialet som bifogas denna rapport.

# Bilaga: Kod för dataanalys

```python
import pandas as pd
import scipy.stats
import matplotlib.cm
import matplotlib.pyplot as plt
import pathlib
```

```python
# Load data

DATA_PATH = "data.xlsx"

NAME = "Namn"
NUTS_SHEET = "Nötter"
LETTER_CODE = "Bokstavskod"
NUM_CODE_SHEET = "Sifferkoder"
NUM_CODE = "Sifferkod"
EVALUATION_SHEET  = "Utvärdering"
TESTER = "Testare"
NUT_TYPE = "Nöttyp"
PROP_VARS = [
    "Flottig",
    "Knaprig",
    "Rostad",
    "Salt",
]
SCORE_VAR = "Betyg"

# Nut pseudonyms

# Nuts were pseudonymized in two steps:
# Person 1 arbitrarily renamed the real names to letter pseudonymes CA, CB, CC, CD, etc.
# Person 2 arbitrarily renamed the letter pseudonyms to C1, C2, C3, C4, etc.
# In combination these two steps renamed nuts to numeric pseudonyms C1, C2, C3, etc.,
# so that neither person knew the name/numeric translation.

name_to_letter = pd.read_excel(DATA_PATH, sheet_name=NUTS_SHEET)
letter_to_num = pd.read_excel(DATA_PATH, sheet_name=NUM_CODE_SHEET)
num_to_name = name_to_letter.merge(letter_to_num, on=LETTER_CODE).set_index(NUM_CODE)[NAME].sort_index()

# Load nut data
nut_data = (
    pd.read_excel(DATA_PATH, sheet_name=NUTS_SHEET)
    .merge(letter_to_num, on=LETTER_CODE)
    .set_index(NUM_CODE)
    .sort_index()
)
nut_data
```

```python
evaluation_data = pd.read_excel(DATA_PATH, sheet_name=EVALUATION_SHEET).set_index(
    [TESTER, NUM_CODE]
)
evaluation_data
```

```python
# Nut types

nut_types = list(nut_data[NUT_TYPE].unique())
nuts_by_type = {
    nut_type: list(nut_data.loc[lambda d: d[NUT_TYPE] == nut_type].index)
    for nut_type in nut_types
}
nuts_by_type
```

```python
# Extract the quantitative data

quant_data = evaluation_data[[*PROP_VARS, SCORE_VAR]].dropna(how="all")
quant_data
```

```python
# Generate figures

NUT_STYLES = [
    dict(marker=marker, color=color)
    for marker, color in zip("vsdpo>", matplotlib.colormaps["tab10"].colors)
]


def num_responses_fig(quant_data, nut_type):
    fig, ax = plt.subplots(figsize=(8, 6))

    counts = (
        quant_data.reindex(nuts_by_type[nut_type], level=NUM_CODE)
        .groupby(NUM_CODE)
        .count()
        .sort_index(ascending=False)
        .rename(lambda num_code: f"{num_to_name[num_code]} ({num_code})")
        .rename_axis("Namn")
    )
    counts.plot.barh(ax=ax)

    ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
    ax.set_xlabel("Antal svar")

    fig.suptitle("Antal svar\n" + nut_type)

    return fig


def score_prediction_fig(quant_data, nut_type):
    quant_data = quant_data.reindex(nuts_by_type[nut_type], level=NUM_CODE)
    nut_means = quant_data.groupby(NUM_CODE).mean()
    nut_stderrs = (
        quant_data.groupby(NUM_CODE).std() / quant_data.groupby(NUM_CODE).count() ** 0.5
    )

    fig, axs = plt.subplots(ncols=len(PROP_VARS), sharey=True, figsize=(9, 4))

    for ax, prop_var in zip(axs, PROP_VARS):
        ax_data = pd.DataFrame(
            {
                "x": nut_means[prop_var],
                "y": nut_means[SCORE_VAR],
                # "xerr": nut_stderrs[prop_var] * 1.96,
                "yerr": nut_stderrs[SCORE_VAR] * 1.96,
            }
        )

        for num_code, style in zip(ax_data.index.unique(NUM_CODE), NUT_STYLES):
            ax.errorbar(
                **ax_data.xs(num_code),
                lw=0,
                elinewidth=1,
                capsize=5,
                **style,
                label=f"{num_to_name[num_code]} ({num_code})",
            )

        ax.set_xlabel(prop_var)

        regression = scipy.stats.linregress(ax_data["x"], ax_data["y"])
        x_pred = ax_data["x"].sort_values()
        y_pred = regression.intercept + x_pred * regression.slope
        ax.plot(x_pred, y_pred, lw=1, ls="--", color="k")

        ax.text(
            0.05,
            0.98,
            f"p = {regression.pvalue:.2f}",
            transform=ax.transAxes,
            verticalalignment="top",
        )

    axs[0].set_ylabel(SCORE_VAR)

    axs[0].legend(loc="upper left", bbox_to_anchor=(0, -0.2), ncols=3)

    fig.tight_layout()

    fig.suptitle(nut_type)

    return fig


figdir = pathlib.Path("figs")
figdir.mkdir(exist_ok=True)

for nt in nut_types:
    fig = num_responses_fig(quant_data, nt)
    fig.savefig(figdir / f"counts-{nt}.png", dpi=100, bbox_inches="tight")

for nt in nut_types:
    fig = score_prediction_fig(quant_data, nt)
    fig.savefig(figdir / f"predictors-{nt}.png", dpi=100, bbox_inches="tight")

for nt in nut_types:
    fig = score_prediction_fig(quant_data, nt)
    fig.savefig(figdir / f"predictors-{nt}.pdf", bbox_inches="tight")
```
